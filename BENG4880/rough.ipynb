```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 20 / (0.5+np.exp(-0.5*(x-10)))

xs = np.linspace(0,20,100)
ys = [f(x) for x in xs]

fig, ax = plt.subplots()

ax.plot(xs, ys, color='k')
ax.hlines(30, 0, 20, linestyles='--', colors='k')
ax.hlines(10, 0, 20, linestyles='--', colors='k')
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.set_ylabel('Biomass')
ax.set_xlabel('Time')
plt.show()
```
